```python
import os
import pandas as pd

# Definir el directorio donde se encuentran los archivos
folder_path = '/content/drive/MyDrive/earth_engine/earth_engine_topo_ai'

# Listar los archivos en la carpeta
file_list = os.listdir(folder_path)

# Filtrar los archivos que correspondan a los archivos .csv de temperatura
csv_files = [file for file in file_list if file.startswith('Temperature_Metropolitana') and file.endswith('.csv')]

# Crear un diccionario para almacenar los archivos por año
data_by_year = {}

# Leer y combinar archivos por año
for file in csv_files:
    # Extraer el año del nombre del archivo
    year = file.split('_')[2]

    # Leer el archivo csv
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)

    # Agregar el archivo al diccionario correspondiente a su año
    if year in data_by_year:
        data_by_year[year].append(df)
    else:
        data_by_year[year] = [df]

# Combinar los archivos por año y remover duplicados
combined_data = {}
for year, dfs in data_by_year.items():
    combined_df = pd.concat(dfs)  # Combinar archivos del mismo año
    combined_df.drop_duplicates(inplace=True)  # Remover duplicados
    combined_data[year] = combined_df  # Guardar el DataFrame combinado

# Mostrar los datos combinados por año (puedes guardarlos si lo prefieres)
for year, df in combined_data.items():
    print(f"Datos combinados para el año {year}:")
    print(df.head())  # Ver las primeras filas de cada año
```

```python
import os
import pandas as pd

# Definir el directorio donde se encuentran los archivos
folder_path = '/content/drive/MyDrive/earth_engine/earth_engine_topo_ai'

# Listar los archivos en la carpeta y filtrar los que comienzan con "Temperature_Metropolitana"
file_list = os.listdir(folder_path)
csv_files = [file for file in file_list if file.startswith('Temperature_Metropolitana') and file.endswith('.csv')]

# Lista para almacenar todos los DataFrames
all_data = []

# Leer y combinar todos los archivos
for file in csv_files:
    # Leer el archivo csv
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    print(df.shape)
    print(df["date"].unique())
    print(len(df["date"].unique()))
    # Agregar el DataFrame a la lista
    all_data.append(df)

# Combinar todos los DataFrames en uno solo
combined_df = pd.concat(all_data, ignore_index=True)

# Remover duplicados
combined_df.drop_duplicates(inplace=True)

# Guardar el DataFrame combinado en un archivo CSV
output_path = '/content/drive/MyDrive/earth_engine/earth_engine_topo_ai/Combined_Temperature_All_Years.csv'
combined_df.to_csv(output_path, index=False)

print(f"Archivo combinado guardado en {output_path}")
```
