````python
import pandas as pd
from utils.db import SQLClient

sql_client = SQLClient("../../db/topoai.db")
sql_client
````

````python
df_1 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2018_1_10000.csv")
df_2 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2018_2_10000.csv")

df_3 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2019_1_10000.csv")
df_4 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2019_2_10000.csv")

df_5 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2020_1_10000.csv")
df_6 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2020_2_10000.csv")

df_7 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2021_1_10000.csv")
df_8 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2021_2_10000.csv")

df_9 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2022_1_10000.csv")
df_10 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2022_2_10000.csv")

df_11 = pd.read_csv("../data/NDVI_NDWI_Metropolitana_2023_1_10000.csv")

data = pd.concat(
    [
        df_1,
        df_2,
        df_3,
        df_4,
        df_5,
        df_6,
        df_7,
        df_8,
        df_9,
        df_10,
        df_11,
    ]
).reset_index(drop=True)

data
````

````python
# Todos los puntos ya tienen un identificador Ãºnico
# en la columna de `system:index`
data.loc[data["system:index"].str.endswith("_0"), ".geo"].head(10).tolist()
````

````python
# Se crean las columnas de coordenadas de logitud y latitud

# fmt: off
data["coordinates"] = (
    data[".geo"]
    .str.split("[").str[1]
    .str.split("]").str[0]
    .str.split(",")
)
# fmt: on

data = data.drop(columns=[".geo"])

data
````

````python
data["longitude"] = data["coordinates"].str[0].astype(float)
data["latitude"] = data["coordinates"].str[1].astype(float)
data = data.drop(columns=["coordinates"])

data
````

````python
# Preparar el dataframe para insertar en la base de datos
data["system:index"] = data["system:index"].str.split("_").str[-1]

data = data.rename(
    columns={
        "system:index": "point_id",
        "NDVI": "ndvi",
        "NDWI": "ndwi",
    },
)
data["date"] = pd.to_datetime(data["date"]).dt.date
data = data.astype(
    {
        "point_id": "int16",
        "ndvi": "float32",
        "ndwi": "float32",
        # "date": "datetime",
        "longitude": "float32",
        "latitude": "float32",
    }
)

# fmt: off
data = (
    data[[
        "point_id", 
        "date", 
        "ndvi", 
        "ndwi", 
        "longitude", 
        "latitude",
    ]]
    .sort_values(["date", "point_id"])
    .reset_index(drop=True)
)
# fmt: on

data
````

```sql
CREATE TABLE IF NOT EXISTS vegetation_index (
  point_id SMALLINT NOT NULL,
  date DATE NOT NULL,
  ndvi FLOAT(13) NOT NULL,
  ndwi FLOAT(13) NOT NULL,
  longitude FLOAT(13) NOT NULL,
  latitude FLOAT(13) NOT NULL

```
